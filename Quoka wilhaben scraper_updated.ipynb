# The working app

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
driver = webdriver.Chrome("Path of the Webdriver goes HERE")
# go to the website
driver.get("https://www.quoka.de/")

#The empty lists will be appended by the loops
df = pd.DataFrame( columns = ['Titles','Prices'])
# These are the keywords I'm looking for regularly
keywords = ['blackmagic','arri','sony fs']
word_text = []
price_text = []



try:
    for keyword in keywords:
#The loop is looking for the searchbar and the search button. When it finds it it performs a click
#it must wait a couple seconds till the page is loaded after it clicks the button
        searchbar = driver.find_element_by_id("search1")
        searchbar.send_keys(keyword)
        button = driver.find_element_by_id("searchbutton")
        button.submit()
        div = WebDriverWait(driver, 15).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "t-nowrap-overflow"))
        )
#The loop is looking for the class of the post's headline, and the class of their price
        for word in div:
            word_text.append(word.text)
            #prices = WebDriverWait(driver, 10).until(
            #    EC.presence_of_all_elements_located((By.cssSelector, "sem.t-bld.t-l.price"))
            #)
        prices = driver.find_elements_by_css_selector('p.sem.t-bld.t-l.price')
        for price in prices:
            price_text.append(price.text)
# when every keywords has been searched and the data has been appeded to the lists the browser closes            
        clear = driver.find_element_by_id("search1")
        clear.clear()

            
# after the loop is done I assign the columns of the df to the lists we have appended
finally:
    df['Titles'] = word_text
    df['Prices'] = price_text
    print(df)
    driver.close()
        
```
